# **List Comprehensions in Python**

List comprehensions provide a concise way to create lists. They are a powerful, declarative, and highly readable alternative to using loops or functional tools like `map()` and `filter()`.

### **Concept 1: Map vs. List Comprehension**
Below is a comparison of how to transform a list of numbers by squaring each value. We contrast the functional `map()` function with a list comprehension.

```python
lst = [1, 2, -5, 4]

def square(x):
    return x * x

# Functional approach using map()
print("Map approach:", list(map(square, lst)))

# List comprehension approach
print("List comprehension:", [square(num) for num in lst])
```

### **Concept 2: Filter vs. List Comprehension**
Here, we filter out only the odd numbers from a list. We compare the functional `filter()` function with an conditional list comprehension.

```python
def is_odd(x):
    return x % 2 == 1

# Functional approach using filter()
print("Filter approach:", list(filter(is_odd, lst)))

# List comprehension approach with condition
print("List comprehension:", [num for num in lst if is_odd(num)])
```

### **Concept 3: Combining Map and Filter vs. List Comprehension**
This example shows how to filter even numbers and square them. It demonstrates how list comprehensions can replace nested `map()` and `filter()` combinations for better readability.

```python
numbers = [1, 2, 3, 4, 5, 6]

# Functional approach using map & filter combined
squared_evens_func = list(map(lambda x: x ** 2, filter(lambda x: x % 2 == 0, numbers)))
print("Functional:", squared_evens_func)

# List comprehension approach
squared_evens_comp = [x ** 2 for x in numbers if x % 2 == 0]
print("List comprehension:", squared_evens_comp)
```

### **Exercise 1: Inline Lambda in List Comprehensions**
You can execute anonymous `lambda` functions inline inside a list comprehension.

```python
# Apply a squaring lambda inline to a list and find the minimum value
min_squared = min([(lambda x: x * x)(x) for x in [1, 2, -5, 4]])
print("Min squared:", min_squared)

# Check if numbers in a list are odd using an inline lambda
is_odd_flags = [(lambda x: x % 2 == 1)(num) for num in lst]
print("Odd flags:", is_odd_flags)
```

### **Exercise 2: Custom Sorting and Min/Max Keys**
Find the element with the maximum absolute/squared value using a custom lambda key.

```python
lst = [1, 2, -5, 4]
max_by_square = max(lst, key=lambda x: x * x)
print("Max by square value:", max_by_square)
```

### **Exercise 3: Enumerating Elements in List Comprehension**
Convert a list of names into a list of tuples containing the index and the name.

```python
names = ['Jerry', 'Kramer', 'Elaine', 'George', 'Newman']
enumerated_names = [(i, name) for i, name in enumerate(names)]
print(enumerated_names)
```

### **Exercise 4: Substring Filtering**
Filter a list of fruits to only keep the ones that contain the letter "a".

```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
fruits_with_a = [fruit for fruit in fruits if "a" in fruit]
print(fruits_with_a)
```

### **Exercise 5: Cubic Range Generation**
Create a list containing the cubes of numbers from 1 to 10.

```python
cubes = [i**3 for i in range(1, 11)]
print(cubes)
```

### **Exercise 6: Lowercasing Strings**
Transform a list of name strings into all lowercase names.

```python
names = ['Jerry', 'Kramer', 'Elaine', 'George', 'Newman']
lowercase_names = [name.lower() for name in names]
print(lowercase_names)
```

### **Exercise 7: Filtering Even Numbers in a Range**
Generate a list of even numbers starting from 2 up to 12.

```python
evens_in_range = [i for i in range(2, 13) if i % 2 == 0]
print(evens_in_range)
```

### **Exercise 8: Flattening a Nested List (List of Lists)**
Flatten a list of lists into a single flat list.

```python
my_list = [[1, 2], [3, 4], [5, 6], [7, 8]]
flattened = [item for sublist in my_list for item in sublist]
print(flattened)
```
