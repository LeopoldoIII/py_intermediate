# **Reduce in Python**

The `reduce()` function from `functools` applies a function cumulatively to the items of a sequence, reducing it to a single value. While Python often prefers built-in functions (`sum`, `max`, `join`) paired with list/generator comprehensions, `reduce()` is key to understanding functional aggregations.

### **Exercise 1: Sum of Numbers**
Calculate the sum of a list of numbers.

**Compare:** The functional `reduce()` approach vs. a list comprehension paired with the built-in `sum()` function.

```python
from functools import reduce

numbers = [1, 2, 3, 4, 5]

# 1. Solution using reduce()
sum_reduce = reduce(lambda x, y: x + y, numbers)
print("Reduce Sum:", sum_reduce)

# 2. Solution using List Comprehension + sum()
sum_comp = sum([num for num in numbers])
print("List Comprehension Sum:", sum_comp)
```

### **Exercise 2: Find the Maximum Number**
Find the maximum number in a list of integers.

**Compare:** The functional `reduce()` approach (comparing elements inline) vs. a list comprehension paired with the built-in `max()` function.

```python
from functools import reduce

numbers = [1, 2, 3, 4, 5]

# 1. Solution using reduce()
max_reduce = reduce(lambda x, y: x if x > y else y, numbers)
print("Reduce Max:", max_reduce)

# 2. Solution using List Comprehension + max()
max_comp = max([num for num in numbers])
print("List Comprehension Max:", max_comp)
```

### **Exercise 3: Concatenate a List of Strings**
Concatenate a list of strings into a single sentence separated by spaces.

**Compare:** The functional `reduce()` approach (joining with string addition) vs. a list comprehension paired with the built-in `.join()` function.

```python
from functools import reduce

words = ["Hello", "world", "this", "is", "Python"]

# 1. Solution using reduce()
concat_reduce = reduce(lambda x, y: x + " " + y, words)
print("Reduce Concatenation:", concat_reduce)

# 2. Solution using List Comprehension + join()
concat_comp = " ".join([word for word in words])
print("List Comprehension Concatenation:", concat_comp)
```

### **Exercise 4: Find the Longest Word**
Find the longest word in a list of strings.

**Compare:** The functional `reduce()` approach (comparing lengths inline) vs. a list comprehension paired with the built-in `max()` function using the `len` key.

```python
from functools import reduce

words = ["Hello", "world", "this", "is", "Python"]

# 1. Solution using reduce()
longest_reduce = reduce(lambda x, y: x if len(x) > len(y) else y, words)
print("Reduce Longest:", longest_reduce)

# 2. Solution using List Comprehension + max(key=len)
longest_comp = max([word for word in words], key=len)
print("List Comprehension Longest:", longest_comp)
```
