# **Lambda Functions in Python**

A lambda function is an anonymous function defined without a name. They are used for creating small, one-line throwaway functions.

### **Concept 1: Basic Anonymous Function**
Lambda functions are defined using the `lambda` keyword. They can take any number of arguments but can only have one expression. Below, we define a lambda function that adds `1` to `x` and call it immediately with the argument `2`.

```python
(lambda x: x + 1)(2)
```

# **lambda & map**

The `map()` function applies a given function to all items in an input list (or any iterable).

### **Concept 2: Squaring Elements and Mapping Multiple Lists**
Below we compare squaring elements using a loop versus using `map()` with a `lambda` function. Additionally, we use a `lambda` that accepts two arguments to sum elements from two lists element-wise.

```python
lst = [1, 2, 4, 7]

# Squaring using a manual loop
result = []
for e in lst:
    result.append(e ** 2)
print("Manual Loop Squaring:", result)

# Squaring using map() and lambda
print("Map & Lambda Squaring:", list(map(lambda x: x ** 2, lst)))

# Element-wise summation of two lists using map() and lambda
lst1 = [1, 2, 4, 7]
lst2 = [1, 2, 4, 7]
print("Element-wise sum:", list(map(lambda x, y: x + y, lst1, lst2)))
```

# **lambda & filter**

The `filter()` function constructs an iterator from elements of an iterable for which a function returns true.

### **Concept 3: Selecting Even Numbers**
We compare selecting even numbers from a list using an imperative loop vs. using `filter()` combined with a `lambda` function.

```python
values = [1, 2, 3, 4, 5, 6, 7, 8, 9]

# Imperative loop approach
pares = []
for value in values:
    if value % 2 == 0:
        pares.append(value)
print("Manual loop filtering:", pares)

# Functional filter() and lambda approach
print("Filter & Lambda filtering:", list(filter(lambda x: x % 2 == 0, values)))
```

# **lambda & reduce**

The `reduce()` function applies a function cumulatively to the items of a sequence to reduce the sequence to a single value.

### **Concept 4: Summing a List**
We calculate the sum of all elements in a list, comparing a manual loop to `reduce()` from the `functools` module.

```python
from functools import reduce

values = [2, 4, 6, 5, 4]

# Manual sum loop
result = 0
for i in values:
    result += i
print("Manual sum:", result)

# Functional reduce() and lambda approach
print("Reduce & Lambda sum:", reduce(lambda x, y: x + y, values))
```

# **lambda & sorted**

Python's `sorted()` function takes a custom sorting key using the `key` parameter, which is commonly specified using a `lambda` function.

### **Concept 5: Lexicographical vs. Custom Numeric Sorting**
When strings contain both characters and numbers (e.g., `'id1'`, `'id30'`, `'id2'`), default alphabetical sorting performs poorly (e.g., `'id100'` gets placed before `'id2'`). A `lambda` function can extract the numeric part so we can sort them numerically.

```python
ids = ['id1', 'id2', 'id30', 'id3', 'id22', 'id100']

# Default Lexicographical sort
print("Default sort:", sorted(ids))
```

```python
# Custom sort using lambda key to cast suffix to integer
print("Custom numeric sort:", sorted(ids, key=lambda x: int(x[2:])))
```

### **Concept 6: Suffix Extraction**
Extracting the raw numeric suffix from the ID strings using a list comprehension.

```python
suffixes = [id_str[2:] for id_str in ids]
print("Suffixes:", suffixes)
```
